> **Solution:**
>- $\underline{\text{Get Forward Price:}}$
>>- Let $F_{t}$ be the time-$t$ forward price for $T_{2}$-delivery of some arbitrary underlying $S$, and let $r$ be the constant interest rate
>>- The forward price $F_{t}$, of a forward contract, paying $S_{T_{2}} - F_{t}$ at time $T_{2}$, has time-$t$ value $0$ <br><br>
$$S_{t} - F_{t}e^{-r \left(T_{2} - t\right)} = 0$$

>>- Long $\left(K, T_{2}\right)$-forward contract, this has a time-$t$ payout of: <br><br>
$$f^{1}_{t} = S_{t} - Ke^{-r \left(T_{2} - t\right)}$$
>> <br>
>>- Short $\left(F_{t}, T_{2}\right)$-forward contract, this has a time-$t$ payout of: <br><br>
$$f^{2}_{t} = S_{t} - F_{t}e^{-r \left(T_{2} - t\right)}$$
>> <br>
>>- The value of the portfolio at time-$t$ ($V_{t} = f_{t} - f_{0} = f_{t}$) is then: <br><br>
$$\begin{align*}
 V_{t} &= f_{t} - f_{0} \\
 &= f_{t} - 0 \\
 &= f^{1}_{t} - f^{2}_{t}\\
 &= \left(S_{t} - Ke^{-r \left(T_{2} - t\right)}\right) - \left(S_{t} - F_{t}e^{-r \left(T_{2} - t\right)}\right) \\

$$\therefore \boxed{f_{t} = \left(F_{t} - K\right)e^{-r \left(T_{2} - t\right)}} \quad \text{Time-t Value of Forward Contract with strike K, Forward Price }F_{t} \text{ and } T_{2}\text{-Delivery}$$


## Problem 1B


> If $S$ is a $\textit{stock}$ paying no dividends, the forward price is $F_{t} = S_{t}e^{r \left(T_{2} - t\right)}$; otherwise arbitrage would exist.
>> If, say, $F_{t} > S_{t}e^{r \left(T_{2} - t\right)}$, then arbitrage would exist: at time-$t$, borrow $S_{t}$ dollars, buy the stock, and short the forward (with delivery price $F_{t}$ and time-$t$ value $0$). At time-$T_{2}$, deliver the stock, and receive $F_{t}$, which is more than enough to cover your accumulated debt of $S_{t}e^{r \left(T_{2} - t\right)}$ dollars.
>
> However, if $S$ is the spot price of a barrel of crude oil (so, for all $t$, the time-$t$ price for time-$t$ delivery is $S_{t}$ per barrel), then this argument fails. Explain briefly (one or two sentences, no math) why $\textit{this specific arbitrage}$ does not apply to crude oil, by specifically pinpointing, in the quote above, why we cannot simply replace "stock" with "crude oil".
> <br><br>
> Hint: Considerpractical complications.



> **Solution:**
>- If $S$ is a stock paying no dividends, the forward price is $F_{t} = S_{t}e^{r \left(T_{2} - t\right)}$; otherwise arbitrage would exist
>- Now, in the case of $S$ being the spot price of a barrel of crude oil, then we have additional cost to consider such as storage and transportation
>>- We need to consider the **Cost of Carry**
>>- This includes fee from warehouse space, utilities, and maintenance
> 
>- Then the price to cover will be $\left(S_{t}e^{r \left(T_{2} - t\right)} + \text{Storage Cost}\right)$, which the forward price $F_{t}$ may not be able to cover since for an arbitrage, in the example, $F_{t} > S_{t}e^{r \left(T_{2} - t\right)}$ and we have to now consider $F_{t} > \left(S_{t}e^{r \left(T_{2} - t\right)} + \text{Storage Cost}\right)$
> <br>
>- The general formula for the forward price, which can be modified for discrete and continuous cases, is the following: <br><br>
$$F_{t} = \begin{cases} 
 S_{t}e^{\left(r - q\right) \left(T_{2} - t\right)} & \text{for dividend-paying stock} \\ 
 S_{t}e^{\left(r + u\right) \left(T_{2} - t\right)} & \text{for commodities with storage cost} \\
\end{cases}$$


## Problem 1C,D


> So we need more assumptions to relate $F_{t}$ and $S_{t}$ (here and in (c,d,e,f,g), the $S$ denotes spot crude oil, and $F_{t}$ denotes the time-$t$ forward price for $T_{2}$-delivery crude oil). One approach is to assume a specific model of the risk-neutral dynamics of $S$. For instance, let us assume that $S$ satisfies <br><br>
$$\begin{align*}
 S_{t} &= \exp{\left(X_{t}\right)} \\
 \mathrm{d}X_{t} &= \kappa\left(\alpha - X_{t}\right)\mathrm{d}t + \sigma\mathrm{d}W_{t}
\end{align*}$$
>
> where $W$ is Brownian motion, under risk-neutral measure.
> <br>
> Then, since $r$ is constant and $\mathbb{E}_{t}\left(e^{-r \left(T_{2} - t\right)}\left(S_{T_{2}} - F_{t}\right)\right)$ must be $0$, one can calculate <br><br>
$$F_{t} = \mathbb{E}_{t}\left(S_{T_{2}}\right) = \exp\left[\log{\left(S_{t}\right)}e^{-\kappa\left(T_{2} - t\right)} + \alpha\left(1 - e^{-\kappa\left(T_{2} - t\right)}\right) + \frac{\sigma^{2}}{4\kappa}\left(1 - e^{-2\kappa\left(T_{2} - t\right)}\right)\right]$$
> <br>
> where $\mathbb{E}_{t}$ is time-$t$ conditional expectation. Suppose $\kappa = 0.472$, $\alpha = 4.4$, $\sigma = 0.368$, $r = 0.05$, and the time-$0$ spot price is $S_{0} = 106.9$.
>
> Let $C$ be the time-$0$ price of a $K$-strike $T_{1}$-expiry European call on $F$. So this call pays $\left(F_{T_{1}} - K\right)^{+}$. Let the call option have strike $K = 103.2$ and expiration $T_{1} = 0.5$. Let the forward have delivery date $T_{2} = 0.75$. See the $\text{ipynb}$ file.


```python
%%time
####================================================
### Model Dynamics (Exponential Ornstein-Uhlenbeck process)
class XOU:
    
    def __init__(self, 
                 kappa: float, 
                 alpha: float, 
                 sigma: float, 
                 S0: float, 
                 r: float) -> None:
        
        self.kappa = kappa
        self.alpha = alpha
        self.sigma = sigma
        self.S0 = S0
        self.r = r
        
        return None


####================================================
### Option Contract
class CallOnForwardPrice:
    
    def __init__(self, 
                 K1: float, 
                 T1: float, 
                 T2: float) -> None:
        
        self.K1 = K1   ### Call option strike
        self.T1 = T1   ### Call option expiration
        self.T2 = T2   ### Forward contract delivery/expiration
        
        return None

```

```python
%%time
####================================================
### Pricing Engine
class MCengine:
    
    def __init__(self, 
                 N: float, 
                 M: float, 
                 epsilon: float, 
                 seed: float = 0) -> None:
        
        self.N = N               ### Number of timesteps on each path
        self.M = M               ### Number of paths
        self.epsilon = epsilon   ### For the dC/dS calculation
        self.rng = np.random.default_rng(seed=seed) 
        ### Seeding the random number generator with a specified number helps make the calculations reproducible
        
        return None
    
    def price_call_XOU(self, 
                       contract: CallOnForwardPrice, 
                       dynamics: XOU) -> Tuple[float]:
        
        ### Seed the random number generator
        self.rng.normal() 
        
        ### Separate dynamics and contract class attributes to variables
        kappa, alpha, sigma, S0, r = dynamics.__dict__.values()
        K1, T1, T2 = contract.__dict__.values()
        
        ### Make timestep
        ### NOTE: The contract expires at T1, so we only need to simulate up to T1
        deltat = T1 / self.N
        
        ### HELPER FUNCTION ###
        ### Function to simulate the path
        def path(S: float, Zt: np.ndarray) -> List[float]:
            ### Make Brownian motion (Wt), and initialize log process
            Wt = np.sqrt(deltat) * Zt             ### Make Brownian Motion
            Xt = np.zeros((self.N + 1, self.M))
            Xt[0] = np.log(S)                     ### Initialize price to log-space 
            
            ### Iterate through all timesteps (up to T1)
            for n in range(1, self.N + 1):
                ### Update next Xt position based on dynamics
                Xt[n] = Xt[n-1] + kappa * (alpha - Xt[n-1]) * deltat + sigma * Wt[n-1]
            
            ### Get the time-T1 forward price for T2-delivery crude oil and call payoff
            tau = T2 - T1  ### This is T2 - t, but since we iterate all the way to N we get t = T1
            Ft = np.exp((np.exp(-kappa * tau) * Xt[-1]) + 
                        (alpha * (1 - np.exp(-kappa * tau))) +
                        (((sigma ** 2) / (4 * kappa)) * (1 - np.exp(-2 * kappa * tau)))
                       )
            call_payoff = np.maximum(Ft - K1, 0)   ### This is at t = T1, Size = M
            
            ### Get time-0 option price and standard error (sample error, M - 1)
            Y0_paths = np.exp(-r * T1) * call_payoff   ### Size = M
            Y0 = np.mean(Y0_paths)
            Y0_std_error = np.std(Y0_paths, ddof=1) / np.sqrt(self.M)
            
            return [Y0, Y0_std_error]
        
        ### Generates pseudo-random normals (Zt) and S0_bump
        Zt = self.rng.normal(loc=0.0, scale=1.0, size=(self.N, self.M))   ### IID random normals
        S0_bump = S0 + self.epsilon
        
        ### Calculate call price (1C), standard error (1C), and call delta (1D)
        call_price, standard_error = path(S=S0, Zt=Zt)                  ### Problem 1C
        call_price_bump, standard_error_bump = path(S=S0_bump, Zt=Zt)   ### We pass the same Zt here
        call_delta = (call_price_bump - call_price) / self.epsilon      ### Problem 1D
        
        return (call_price, standard_error, call_delta)

```

## Problem 1C


> Estimate $C\left(S_{0}\right)$ using Monte Carlo simulation of $S$ with $100$ timesteps on $\left[0, T_{1}\right]$. Choose the number of paths large enough that the standard error $[\text{the sample standard deviation, divided by the square root of the number of paths}]$ is less than $0.05$. Report the standard error. Don't use any variance reduction technique.



> **Solutions**
>- $\underline{\text{Spot Crude Oil Dynamics:}}$
>>- Note: The general basis for the formulation below are from the **Vasicek** and **CIR** models
>>- We will simulate the following $\text{SDE}$: <br><br>
$$\begin{align*}
 S_{t} &= \exp{\left(X_{t}\right)} \\
 \mathrm{d}X_{t} &= \kappa\left(\alpha - X_{t}\right)\mathrm{d}t + \sigma\mathrm{d}W_{t}
\end{align*}$$
>> <br>
>>- If we apply $\text{Ito}$'s rule to $e^{\kappa t}X_{t}$, then we can get: <br><br>
$$X_{T} \thicksim \mathcal{N}\left(X_{t}e^{-\kappa\left(T - t\right)} + \alpha\left(1 - e^{-\kappa\left(T - t\right)}\right), \frac{\sigma^{2}}{2\kappa}\left(1 - e^{-2\kappa\left(T - t\right)}\right)\right)$$
>>>- Proof:
>>>>- Ornstein-Uhlenbeck process has a well-known closed-form solution: <br><br>
$$X_{T} = X_{t}e^{-\kappa\left(T - t\right)} + \alpha\left(1 - e^{-\kappa\left(T - t\right)}\right) + \sigma\int^{T}_{t}e^{-\kappa\left(T - u\right)}\mathrm{d}W_{u}$$
>>>>>- We can see that $X_{T} \thicksim \mathcal{N}\left(\mathbb{E}_{t}\left(X_{T}\right), \text{Var}_{t}\left(X_{T}\right)\right)$
>>>> <br>
>>>>- The time-$t$ conditional expectation of $X_{T}$ is: <br><br>
$$\mathbb{E}_{t}\left(X_{T}\right) = X_{t}e^{-\kappa\left(T - t\right)} + \alpha\left(1 - e^{-\kappa\left(T - t\right)}\right)$$
>>>> <br>
>>>>- The time-$t$ conditional variance of $X_{T}$ is: <br><br>
$$\begin{align*}
 \text{Var}_{t}\left(X_{T}\right) &= \mathbb{E}_{t}\left[\left(X_{T} - \mathbb{E}\left[X_{T}\right]\right)^{2}\right] \\
 &= \text{Var}\left(\sigma\int^{T}_{t}e^{-\kappa\left(T - u\right)}\mathrm{d}W_{u}\right) \\
 &= \sigma^{2}\int^{T}_{t}e^{-2\kappa\left(T - u\right)}\mathrm{d}u \\
 &= \frac{\sigma^{2}}{2\kappa}\left(1 - e^{-2\kappa\left(T - t\right)}\right)
\end{align*}$$
>>>> <br>
>>>>- We can solve for the time-$t$ conditional mean and variance for $S_{T}$ now by using the following: <br><br>
$$\mathbb{E}_{t}\left(S^{n}_{T}\right) = e^{n\mathbb{E}_{t}\left(X_{T}\right) + \frac{n^{2}}{2}\text{Var}_{t}\left(X_{T}\right)}$$
>>>> <br>
>>>>- The time-$t$ conditional mean of $S_{T}$ is: <br><br>
$$\begin{align*}
 \mathbb{E}_{t}\left(S_{T}\right) &= e^{\mathbb{E}_{t}\left(X_{T}\right) + \frac{1}{2}\text{Var}_{t}\left(X_{T}\right)} \\
 &= e^{\left[\log{\left(S_{t}\right)}e^{-\kappa\left(T_{2} - t\right)} + \alpha\left(1 - e^{-\kappa\left(T_{2} - t\right)}\right) + \frac{\sigma^{2}}{4\kappa}\left(1 - e^{-2\kappa\left(T_{2} - t\right)}\right)\right]}
\end{align*}$$
>>>> <br>
>>>>- The time-$t$ conditional variance of $S_{T}$ is: <br><br>
$$\begin{align*}
 \text{Var}_{t}\left(S_{T}\right) &= \mathbb{E}_{t}\left[\left(S_{T} - \mathbb{E}\left[S_{T}\right]\right)^{2}\right] \\
 &= \mathbb{E}_{t}\left[S^{2}_{T}\right] - \left(\mathbb{E}\left[S_{T}\right]\right)^{2} \\
 &= e^{2\mathbb{E}_{t}\left(X_{T}\right) + \text{Var}_{t}\left(X_{T}\right)}\left(e^{\text{Var}_{t}\left(X_{T}\right)} - 1\right) \\
 &= e^{\left[2\log{\left(S_{t}\right)}e^{-\kappa\left(T_{2} - t\right)} + 2\alpha\left(1 - e^{-\kappa\left(T_{2} - t\right)}\right) + \frac{\sigma^{2}}{2\kappa}\left(1 - e^{-2\kappa\left(T_{2} - t\right)}\right)\right]}\left(e^{ \frac{\sigma^{2}}{2\kappa}\left(1 - e^{-2\kappa\left(T - t\right)}\right)} - 1\right)
\end{align*}$$
>>
>>- The time-$t$ forward price for $T_{2}$-delivery crude oil as: <br><br>
$$\begin{align*}
 F_{t} &= \mathbb{E}_{t}\left(S_{T_{2}}\right) = \exp\left[\log{\left(S_{t}\right)}e^{-\kappa\left(T_{2} - t\right)} + \alpha\left(1 - e^{-\kappa\left(T_{2} - t\right)}\right) + \frac{\sigma^{2}}{4\kappa}\left(1 - e^{-2\kappa\left(T_{2} - t\right)}\right)\right] \\
 &= \mathbb{E}_{t}\left(e^{X_{T_{2}}}\right) = \exp\left[X_{t}e^{-\kappa\left(T_{2} - t\right)} + \alpha\left(1 - e^{-\kappa\left(T_{2} - t\right)}\right) + \frac{\sigma^{2}}{4\kappa}\left(1 - e^{-2\kappa\left(T_{2} - t\right)}\right)\right]
\end{align*}$$
>> <br>
>>- Here we let $C$ be the time-$0$ price of a $K$-strike $T_{1}$-expiry European call on $F_{t}$. 
>>>- So this call pays at time-$T_{1}$:
$$C = \left(F_{T_{1}} - K\right)^{+}$$
>>> <br>
>>>- Time time-$0$ price is computed by:
$$\begin{align*}
 C_{0} &= e^{-r T_{1}}\mathbb{E}_{0}\left[C_{T_{1}}\right]\\
 &= e^{-r T_{1}}\mathbb{E}_{0}\left[\left(F_{T_{1}} - K\right)^{+}\right]\\
\end{align*}$$
> <br>
>- $\underline{\text{Monte Carlo Algorithm:}}$
>>- Make Brownian motion dynamics used to update path steps <br><br>
$$\begin{align*}
 \Delta t &= T_{1} / N \\
 Z_{t} &= \mathcal{N}\left(\mu = 0, \text{var} = 1, \text{size} = \left(N, M\right)\right) \\
 W_{t} &= \sqrt{\Delta t} \cdot Z_{t}
\end{align*}$$
>> 
>>- Path Update
>>>- 1. Initialize $X_{0} = \log\left(S_{0}\right)$
>>>- 2. Store the dynamics $X_{T} = \left[X_{0}\right]$
>>>- 3. Iterate over all timestep $\left[0, T_{1}\right]$ to traverse the path, and update $X_{t}$ for each step wrt its dynamics: <br><br>
$$X_{i} = X_{i-1} + \kappa\left(\alpha - X_{i-1}\right)\Delta t + \sigma W_{i-1}$$
>>> <br>
>>>- 4. Calculate time-$t$ forward price for $T_{2}$-delivery crude oil after the full path is traversed: <br><br>
$$F_{t} = \exp\left[X_{t}e^{-\kappa\left(T_{2} - t\right)} + \alpha\left(1 - e^{-\kappa\left(T_{2} - t\right)}\right) + \frac{\sigma^{2}}{4\kappa}\left(1 - e^{-2\kappa\left(T_{2} - t\right)}\right)\right]$$
>>> <br>
>>>- 5. Calculate call payoff: <br><br>
$$C = e^{-r T_{1}}\mathbb{E}_{0}\left[\left(F_{T_{1}} - K\right)^{+}\right]$$
>>>>- Note: we take the mean of $C$ since it's the time-$0$ price over many paths

## Problem 1E


> Calculate analytically $\partial f_{0}/\partial S_{0}$, where $f_{0}$ is the time-$0$ value of a position long one forward contract on a barrel of crude oil, with delivery date $T_{2}$ and some fixed delivery price $K$.



> **Solution:**
>- Let $f_{t}$ be the time-$t$ value of a $T_{2}$-forward contract with delivery price $K$ on the same underlying, then: <br><br>
$$\therefore \boxed{f_{t} = \left(F_{t} - K\right)e^{-r \left(T_{2} - t\right)}} \quad \text{Time-t Value of Forward Contract with strike K, Forward Price }F_{t} \text{ and } T_{2}\text{-Delivery}$$
> <br>
>- From part $\left(\text{c}\right)$, we have the following relations: <br><br>
$$\begin{align*}
 X_{t} &= \log\left(S_{t}\right) \\
 F_{t} &= \mathbb{E}_{t}\left(e^{X_{T_{2}}}\right) = \exp\left[X_{t}e^{-\kappa\left(T_{2} - t\right)} + \alpha\left(1 - e^{-\kappa\left(T_{2} - t\right)}\right) + \frac{\sigma^{2}}{4\kappa}\left(1 - e^{-2\kappa\left(T_{2} - t\right)}\right)\right] \\
 f_{t} &= \left(F_{t} - K\right)e^{-r \left(T_{2} - t\right)}
\end{align*}$$
> <br>
>- From above, we can use related rates from ordinary calculus to get the proper form of $\frac{\partial f_{0}}{\partial S_{0}}$: <br><br>
$$\frac{\partial f_{0}}{\partial S_{0}} = \frac{\partial f_{t}}{\partial F_{t}} \frac{\partial F_{t}}{\partial X_{t}} \frac{\partial X_{t}}{\partial S_{t}}\Bigg|_{t=0}$$
> <br>
>- We have: <br><br>
$$\begin{align*}
 \frac{\partial f_{t}}{\partial F_{t}} &= \frac{\partial}{\partial F_{t}}\left[\left(F_{t} - K\right)e^{-r \left(T_{2} - t\right)}\right] = e^{-r\left(T_{2} - t\right)} \\
 \frac{\partial F_{t}}{\partial X_{t}} &= \frac{\partial}{\partial X_{t}} \left[\mathbb{E}_{t}\left(e^{X_{T_{2}}}\right)\right] = e^{-\kappa\left(T_{2} - t\right)}F_{t} \\
 \frac{\partial X_{t}}{\partial S_{t}} &= \frac{\partial}{\partial S_{t}}\left[\log\left(S_{t}\right)\right] = \frac{1}{S_{t}}
\end{align*}$$
> <br>
>- Therefore, with $r = 0.05$, $\kappa = 0.472$, $T_{2} = 0.75$, $S_{0} = 106.9$, and $F_{0} = 102.23$, we get: <br><br>
$$\begin{align*}
 \frac{\partial f_{0}}{\partial S_{0}} &= \frac{\partial f_{t}}{\partial F_{t}} \frac{\partial F_{t}}{\partial X_{t}} \frac{\partial X_{t}}{\partial S_{t}}\Bigg|_{t=0} \\ 
 &= e^{-\left(r + \kappa\right)\left(T_{2} - t\right)}\frac{F_{t}}{S_{t}}\Bigg|_{t=0} \\
 &= e^{-\left(r + \kappa\right)T_{2}}\frac{F_{0}}{S_{0}} \\
 &= 0.64651
\end{align*}$$


```python
%%time
####================================================
### Function to compute Forward price, forward contract value and partial derivative of forward contract wrt S
def calculate_Ft_ft_del_fS(contract: CallOnForwardPrice, 
                           dynamics: XOU, 
                           t: float = 0) -> List[float]:
    
    ### Separate dynamics and contract class attributes to variables
    K1, T1, T2 = contract.__dict__.values()
    kappa, alpha, sigma, S0, r = dynamics.__dict__.values()
    
    ### Set up the constants to to get forward price
    Xt = np.log(S0)
    tau = T2 - t
    Ft = np.exp((np.exp(-kappa*tau) * Xt) + 
                (alpha * (1 - np.exp(-kappa*tau))) +
                (((sigma**2) / (4*kappa)) * (1 - np.exp(-2*kappa*tau)))
               )
    
    ### Compute derivative of forward contract wrt S
    partial_fS = np.exp(-(r + kappa) * tau) * (Ft / S0)   ### Problem 1E
    
    ### Compute forward contract value
    ft = np.exp(-r * tau) * (Ft - K1)   ### Problem 1G
    
    return [Ft, ft, partial_fS]

```

```python
%%time
####================================================ 
### Compute derivative of forward contract wrt S
F0, f0, partial_fS = calculate_Ft_ft_del_fS(hw6contract, hw6dynamics, t=0)
print('Time-0 Forward Price F0: {:.2f}'.format(F0))
print('Time-0 value of partial_f / partial_S = {:.5f}\n'.format(partial_fS))
```

## Problem 1F


> Suppose you want to hedge a position short one call (so your hedge portfolio should replicate a position long one call), by continuously rebalancing a position in $T_{2}$-delivery forward contracts. Your hedge portfolio at time-$0$ should be long how many forward contracts? Your final answer should be a number. 
> <br><br>
> The delivery price $K$ of the forward contracts is irrelevant to the answer here; it would affect only how many units of the bank account to carry in the portfolio (which I am not asking you to compute).



> **Solution:**
>- You are short one European call option, and you want to hedge that position by building a replicating portfolio using $T_{2}$-delivery forward contracts on the underlying asset
>- If you want to hedge a position short one call you want your portfolio to be composed of $-1$ unit of a call option and $\Delta_{f}$ units of forward contracts
>- We will have our portfolio as $\theta = \left[-1, \Delta_{f}\right]$, and $X_{t} = \left[C_{t}, f_{t}\right]$, where $\theta$ is the quantity and $X_{t}$ are the assets at time-$t$
>- The portfolio value will be: <br><br>
$$\begin{align*}
 V_{t} &= \theta \cdot X_{t} \\
 &= -C_{t} + \Delta_{f} f_{t} \\
\end{align*}$$
> <br>
>- We want this portfolio to be delta hedged, meaning create a delta-neutral portfolio, so the total delta of our portfolio is zero: <br><br>
$$\begin{align*}
 \frac{\partial V_{t}}{\partial S_{t}} &= 0 \\
 &= -\frac{\partial C_{t}}{\partial S_{t}} + \Delta_{f} \frac{\partial f_{t}}{\partial S_{t}} \\
\end{align*}$$
> <br>
>- Therefore: <br><br>
$$\frac{\partial C_{t}}{\partial S_{t}} =  \Delta_{f} \cdot \frac{\partial f_{t}}{\partial S_{t}}$$
> <br>
>- From above we have that: <br><br>
$$\begin{align*}
 \frac{\partial f_{0}}{\partial S_{0}} &= \frac{\partial f_{t}}{\partial F_{t}} \frac{\partial F_{t}}{\partial X_{t}} \frac{\partial X_{t}}{\partial S_{t}}\Bigg|_{t=0} \\
 &= e^{-\left(r + \kappa\right)\left(T_{2} - t\right)}\frac{F_{t}}{S_{t}}\Bigg|_{t=0} \\
\end{align*}$$
> <br>
>- Then the number of forward contract to hedge at time-$0$ is: <br><br>
$$\boxed{\Delta_{f} = \frac{\partial C_{t}/\partial S_{t}}{\partial f_{t}/\partial S_{t}}\Bigg|_{t=0}}$$


```python
%%time
####================================================
### Compute the number of forward contract to hedge at time-0
delta_f = call_delta / partial_fS
print('Time-0 Call Delta: {:.5f}'.format(call_delta))
print('At time-0, the hedge portfolio should be long {:.5f} forward contracts\n'.format(delta_f))
```

## Problem 1G


> Consider the following "purchase agreement" contract. The holder of this contract receives time-$T_{2}$ delivery of $\theta$ barrels of crude oil, and pays, at time-$T_{2}$, a delivery price of $K$ dollars per barrel. The $\theta$ is chosen at time-$T_{1}$ by the holder of the purchase agreement, subject to the restriction that $4000 \leq \theta \leq 5000$; in particular, $\theta = 0$ is not a valid choice, because the contract is a commitment to purchase at least $4000$ barrels. Using your answer to $\left(\text{c}\right)$, without running any new simulations, find the time-$0$ value of this contract. 
> <br><br>
> Here $K, T_{1}, T_{2}$ have the same values as on the previous page.
> <br><br>
> Hint: Assume the holder acts optimally; thus $\theta$ is either $4000$ or $5000$, depending on $F_{T_{1}}$.



> **Solution:**
>- The holder of the purchase agreement contract recieves time-$T_{2}$ delivery of $\theta$ barrels of crude oil, and pays, at time $T_{2}$, $K$ dollars per barrel
>- The contract holder acts optimally, so let $4000 \leq \theta \leq 5000$ with $\theta_{max} = 5000$ and dependent on $F_{T_{1}}$
>>- The holder of the purchase agreement chooses $\theta$ at time $T_{1}$
> 
>- The time-$T_{2}$ value of our portfolio is: <br><br>
$$V_{T_{2}} = \theta\left(F_{T_{2}} - K\right) + \left(\theta_{max} - \theta\right)\left(F_{T_{1}} - K\right)$$
> <br>
>- The portfolio can be replicated by: <br><br>
$$V_{T_{2}} = \theta f_{T_{2}} + \left(\theta_{max} - \theta\right) C_{T_{1}}$$
> <br>
>- Then, the time-$0$ value of our contract is the expectation at time-$0$ for the replication: <br><br>
$$\begin{align*}
 V_{0} &= \mathbb{E}_{0}\left[V_{T_{2}}\right] \\
 &= \theta \mathbb{E}_{0}\left[f_{T_{2}}\right] + \left(\theta_{max}-\theta\right)\mathbb{E}_{0}\left[e^{-r T_{2}}C_{T_{1}}\right] 
\end{align*}$$
> <br>
>- The time-$0$ value of the purchase agreement contract is: <br><br>
$$V_{0} = \theta f_{0} + \left(\theta_{max} - \theta\right) C_{0}e^{-r \left(T_{2} - T_{1}\right)}$$
> <br>
>- With: <br><br>
$$f_{0} = \left(F_{0} - K\right)e^{-r T_{2}}$$
> <br>
>- Therefore, we long $4,000$ forward contracts and long $1,000$ call options


```python
%%time
####================================================
### Compute the forward contract value
K1, T1, T2 = hw6contract.__dict__.values()
kappa, alpha, sigma, S0, r = hw6dynamics.__dict__.values()

### Compute the time-0 purchase agreement contract value
theta, theta_MAX = 4000, 5000
contract_t0 = (theta * f0) + ((theta_MAX - theta) * call_price * np.exp(-r * (T2 - T1)))
print('Time-0 Forward Call Price: {:.5f}'.format(call_price))
print('Time-0 Forward Price F0: {:.5f}'.format(F0))
print('Time-0 Forward Contract Value f0: {:.5f}'.format(f0))
print('Time-0 purchase agreement contract value is: {:.5f}\n'.format(contract_t0))
```

## Problem 2A


> Let $S_{t}$ be the time-$t$ price of a stock that continuously pays a constant proportional dividend yield $q$. This means that each $1$ share of $S$ at time $0$ will grow, via dividend reinvestment, to $e^{qt}$ shares of $S$ at each time $t \geq 0$. In other words, the stock should not be regarded as a tradeable asset. Rather, units of the "bundle" should be regarded as tradeable, where $1$ unit of the bundle is defined to be <br><br>
$$e^{qt} \text{ shares of stock, at all times } t \geq 0$$
> <br>
> Aside from the above information, do not assume any specific dynamics for $S$.
> <br><br>
> Using replication, find the time-$0$ value of a forward contract which pays $S_{T} − K$ at time $T$.
> <br>
> How many shares of $S$ and units of $B$ does the replicating portfolio hold at time $0$?
> <br><br>
> What is the time-$0$ $\textit{forward price}$ for time-$T$ delivery of $S_{T}$? (This is not the same thing as the $\textit{value of a forward contract}$.)



> **Solution:**
>- We need to make a tradeable bundle before constructing the replicating portfolio
>- The tradeable bundle, $X_{t}$, has the following boundary conditions: <br><br>
$$X_{T} = S_{T} \quad\text{&}\quad X_{0} = e^{-qT}S_{0}$$
> <br>
>- Then the time-$t$ tradeable bundle is: <br><br>
$$X_{t} = e^{-q\left(T - t\right)}S_{t}$$
>>- Holding $1$ unit of $X_{t}$ is equivalent to holding $e^{-q\left(T - t\right)}$ unit of $S_{t}$
> <br>
>- Additionally, a bank account is tradeable and can be related to bonds
>- The tradeable zero coupon bond, $Z_{t}$, and bank account , $B_{t}$, have the following boundary conditions: <br><br>
$$\begin{align*}
 Z_{T} &= 1 \quad\text{&}\quad Z_{0} = e^{-rT} \\
 B_{T} &= e^{rT} \quad\text{&}\quad B_{0} = 1
\end{align*}$$
> <br>
>- The time-$t$ tradeable zero coupon bond and bank account have the following relations: <br><br>
$$\begin{align*}
 Z_{t} &= e^{-r\left(T - t\right)} = \frac{1}{B_{T}} B_{t} \\
 B_{t} &= e^{r t} = \frac{1}{Z_{0}} Z_{t}
\end{align*}$$
>>
>> **Forward Contract Value:**
>>- We have a forward contract, $f_{T}$, which has a time-$T$ payout of $S_{T} − K$
>>- We will replicate this using the stock $S$ and bank account $B$
>>>- Can only use tradeable assets so the portfolio is constructed using the tradeable bundle $X$ and bank account $B$
>>>- At the end we will convert this to a portfolio using the stock $S$ and bank account $B$
>>- Our time-$T$ replicating portfolio is the following: <br><br>
$$\begin{align*}
 V_{T} &= f_{T} \\
 &= S_{T} − K \\
 &= X_{T} − K \\
 &= \alpha X_{T} + \beta B_{T}
\end{align*}$$
>> <br>
>>- Solving for $\alpha$ and $\beta$, we get: <br><br>
$$\begin{align*}
 \alpha &= 1 \\
 \beta &= −\frac{K}{B_{T}} = -Ke^{-r T}
\end{align*}$$
>> <br>
>>- Then, the time-$t$ value of our replicating portfolio is: <br><br>
$$\begin{align*}
 V_{t} &= \alpha X_{t} + \beta B_{t} \\
 &= X_{t} + \beta B_{t} \\
 &= X_{t} + \left(-Ke^{-r T}\right) B_{t} \\
 &= e^{-q\left(T - t\right)}S_{t} + \left(-Ke^{-r T}\right) B_{t}
\end{align*}$$
>> <br>
>>- At time-$t$, the number of units of $S_{t}$ and $B_{t}$ to replicate the forward contract and its value is: <br><br>
$$\begin{align*}
 V_{t} &= e^{-q\left(T - t\right)}S_{t} + \left(-Ke^{-r T}\right) B_{t} \\
 \Theta_{t} &= \begin{cases}
 e^{-q\left(T - t\right)} & \text{units of S} \\ 
 -Ke^{-r T} & \text{units of B}
 \end{cases} \\
\end{align*}$$
>> <br>
>>- At time-$0$, the number of units of $S_{0}$ and $B_{0}$ to replicate the forward contract and its value is: <br><br>
$$\begin{align*}
 V_{0} &= e^{-qT}S_{0} + \left(-Ke^{-r T}\right) B_{0} \\
 \Theta_{0} &= \begin{cases}
 e^{-q T} & \text{units of S} \\ 
 -Ke^{-r T} & \text{units of B}
 \end{cases} \\
\end{align*}$$
>> <br>
>>
>> **Forward Price:**
>>- The forward price $F_{t}$ is the value of $K$ that makes $V_{t} = 0$
>>>- Let $K = F_{t}$ be the value where $V_{t} = 0$
>>>- Then: <br><br>
$$\begin{align*}
 V_{t} &= 0 \\
 &= e^{-q\left(T - t\right)}S_{t} + \left(-F_{t}e^{-r T}\right) B_{t} \\
 &= e^{-q\left(T - t\right)}S_{t} + \left(-F_{t}e^{-r \left(T - t\right)}\right)\\
 F_{t}e^{-r\left(T - t\right)} &= e^{-q\left(T - t\right)}S_{t} \\
 F_{t} &= e^{\left(r - q\right)\left(T - t\right)}S_{t}
\end{align*}$$
>> <br>
>>- At time-$t$, the forward price for time-$T$ delivery of $S_{T}$ is: <br><br>
$$F_{t} = S_{t}e^{\left(r - q\right)\left(T - t\right)}$$
>> <br>
>>- At time-$0$, the forward price for time-$T$ delivery of $S_{T}$ is: <br><br>
$$F_{0} = S_{0}e^{\left(r - q\right)T}$$



## Problem 2B


> Let $S_{t}$ be the time-$t$ price of a stock that pays a fixed dollar dividend $D$ discretely at time $T_{0}$, where $0 < T_{0} < T$. Assume $S$ does not pay any other dividends between time $0$ and $T$. 
> <br><br>
> So we will consider as tradeable the following bundle. One unit of the bundle consists of: <br><br>
$$\begin{align*}
 1 \text{ shares } \quad &\text{ at all times } t < T_{0} \\

Unlike $\left(\text{a}\right)$, the bundle in $\left(\text{b}\right)$ allocates the dividend into bank account units, instead of into more stock shares.
> <br><br>
> We are still not assuming any specific dynamics for $S$.
> <br><br>
> Using replication, find the time-$0$ value of a forward contract which pays $S_{T} − K$ at time $T$.
> <br>
> How many shares of $S$ and units of $B$ does the replicating portfolio hold at time $0$?
> <br><br>
> What is the time-$0$ $\textit{forward price}$ for time-$T$ delivery of $S_{T}$? (This is not the same thing as the $\textit{value of a forward contract}$.)



> **Solution:**
>- We need to make a tradeable bundle before constructing the replicating portfolio
>- The time-$t$ tradeable bundle, $X_{t}$, is the following using the given conditions conditions: <br><br>
$$\begin{align*}
 X_{t} &= S_{t} + De^{-r T_{0}} B_{t} \mathbb{1}_{t \geq T_{0}} \\
 &= S_{t} + De^{r\left(t - T_{0}\right)}\mathbb{1}_{t \geq T_{0}}
\end{align*}$$

>> **Forward Contract Value:**
>>- We have a forward contract, $f_{T}$, which has a time-$T$ payout of $S_{T} − K$
>>- We will replicate this using the stock $S$ and bank account $B$
>>>- Can only use tradeable assets so the portfolio is constructed using the tradeable bundle $X$ and bank account $B$
>>>- Then, once we have a form to simplify, we will replace the tradeable bundle with its relationship form

$$\begin{align*}
 V_{T} &= f_{T} \\
 &= S_{T} − K \\
 &= \alpha X_{T} + \beta B_{T} \\
 &= \alpha\left[S_{T} + De^{-r T_{0}} B_{T}\right] + \beta B_{T} \\
 &= \alpha S_{T} + \left[\alpha De^{-r T_{0}} + \beta\right] B_{T}
\end{align*}$$
>> <br>
>>- Solving for $\alpha$ and $\beta$, we get: <br><br>
$$\begin{align*}
 \alpha &= 1 \\
 \beta &= −\frac{1}{B_{T}}\left[\alpha De^{-r T_{0}}B_{T} + K\right] \\
 &= −\frac{1}{B_{T}}\left[De^{-r T_{0}}B_{T} + K\right] \\
 &= −\left[De^{-r T_{0}} + Ke^{-r T}\right]
\end{align*}$$
>> <br>
>>- Then, the time-$t$ value of our replicating portfolio is: <br><br>
$$\begin{align*}
 V_{t} &= \alpha X_{t} + \beta B_{t} \\
 &= X_{t} + \beta B_{t} \\
 &= S_{t} + De^{r\left(t - T_{0}\right)}\mathbb{1}_{t \geq T_{0}} + \left(−\left[De^{-r T_{0}} + Ke^{-r T}\right]\right) B_{t} \\
 &= S_{t} + \left(−\left[De^{-r T_{0}}\left(1 - \mathbb{1}_{t \geq T_{0}}\right) + Ke^{-r T}\right]\right) B_{t}
\end{align*}$$
>> <br>
>>- At time-$t$, the number of units of $S_{t}$ and $B_{t}$ to replicate the forward contract and its value is: <br><br>

 V_{t} &= S_{t} + \left(−\left[De^{-r T_{0}}\left(1 - \mathbb{1}_{t \geq T_{0}}\right) + Ke^{-r T}\right]\right) B_{t} \\
 \Theta_{t} &= \begin{cases}
 1 & \text{units of S} \\ 
 −\left[De^{-r T_{0}}\left(1 - \mathbb{1}_{t \geq T_{0}}\right) + Ke^{-r T}\right] & \text{units of B}
 \end{cases} \\
\end{align*}$$
>> <br>
>>- At time-$0$, the number of units of $S_{0}$ and $B_{0}$ to replicate the forward contract and its value is: <br><br>
$$\begin{align*}
 V_{0} &= S_{0} + \left(−\left[De^{-r T_{0}} + Ke^{-r T}\right]\right) B_{0} \\
 \Theta_{0} &= \begin{cases}
 1 & \text{units of S} \\ 
 −\left[De^{-r T_{0}} + Ke^{-r T}\right] & \text{units of B}
 \end{cases} \\
\end{align*}$$
>> <br>
>>
>> **Forward Price:**
>>- The forward price $F_{t}$ is the value of $K$ that makes $V_{t} = 0$
>>>- Let $K = F_{t}$ be the value where $V_{t} = 0$
>>>- Then: <br><br>
$$\begin{align*}
 V_{t} &= 0 \\
 &= S_{t} + De^{r\left(t - T_{0}\right)}\mathbb{1}_{t \geq T_{0}} + \left(−\left[De^{-r T_{0}} + F_{t}e^{-r T}\right]\right) B_{t} \\
 &= S_{t} + \left(−\left[De^{-r T_{0}}\left(1 - \mathbb{1}_{t \geq T_{0}}\right) + F_{t}e^{-r T}\right]\right) B_{t} \\
 F_{t}e^{-r T}B_{t} &= S_{t} - De^{-r T_{0}}\left(1 - \mathbb{1}_{t \geq T_{0}}\right)B_{t} \\

>>- At time-$t$, the forward price for time-$T$ delivery of $S_{T}$ is: <br><br>
